# 2. Data Cleaning

## a. Manipulasi kolom meliputi penggantian nama kolom, penghapusan kolom, atau penambahan kolom

````python
# menghilangkan spasi yang ada awal dan akhir nama kolom
df.columns= df.columns.str.strip()
````

````python
# menampilkan semua nama kolom
df.columns
````

## b. Penggantian tipe data yang sesuai dengan seharusnya. Untuk nilai harga dengan simbol mata uang, simbol tersebut dapat dibuang supaya dapat terkonversi ke tipe data yang seharusnya. Perhatikan format penulisan angkanya apakah ada tanda yang membuat angka hanya bisa bertipe data string. (Tambahan: Di suatu kolom yang bernilai uang, terpadat beberapa data yang mengandung karakter '(' dan ')' di dalamnya sehingga harus ditake-out karakter tersebut).

## c. Cek dan handling missing value. (Hint: terdapat non-standard missing value pada kolom yang semestinya bertipe data integer, handling dapat dilakukan dengan imputasi dengan angka nol '0')

- format dari setiap kolom masih belum tepat sehingga belum bisa dilakukan data analysis. 
- kolom data `date` masih bertipe obeject, dimana akan lebih sesuai jika formatnya date time.
(dengan melakukan pd.to_datetime)
- kolom `Unit Sold`,`Manufacturing Price`, `Sale Price`,`Gross Sales`,`Discounts`,`Sales`,`COGS`,`Profit` masih bertipe object, dimana kurang tepat, dan lebih tepat jika formatnya int atau float.
(pertama dilakukan penghilangan karaktek '$', ',', ' ', '(', ')', '-'. kemudian mengubah data type dengan as type)

````python
# melakukan replace data menjadi kosong agar
df['Units Sold'] = df['Units Sold'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','')
df['Manufacturing Price'] = df['Manufacturing Price'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','')
df['Sale Price'] = df['Sale Price'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','')
df['Gross Sales'] = df['Gross Sales'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','')
df['Discounts'] = df['Discounts'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','0')
df['Sales'] = df['Sales'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','')
df['COGS'] = df['COGS'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','')
df['Profit']= df['Profit'].str.replace('$','').str.replace(',','').str.replace('(','').str.replace(')','').str.replace(' ','').str.replace('-','0')
````

````python
# menampilkan ringkasan singkat tentang data saya, rangeindex, total kolom, nama kolom, berupa jumlah nilai bukan nol di setiap kolom dan tipe data kolom
df.info()
````

````python
# merubah data type dari object menjadi float
df['Units Sold']  = df['Units Sold'].astype(float)
df['Manufacturing Price']  = df['Manufacturing Price'].astype(float)
df['Sale Price'] = df['Sale Price'].astype(float)
df['Gross Sales'] = df['Gross Sales'].astype(float)
df['Discounts'] = df['Discounts'].astype(float)
df['Sales'] = df['Sales'].astype(float)
df['COGS'] = df['COGS'].astype(float)
df['Profit']= df['Profit'].astype(float)
````

````python
# merubah data type dari object menjadi date
df['Date'] = pd.to_datetime(df['Date'])
df['Month Number'] = pd.to_datetime(df['Month Number'])
````

````python
# Melakukan NF dengan mereplace value menjadi (1,2,3,4,...)
df['Segment'] = df['Segment'].str.replace('Government','1').str.replace('Midmarket','2').str.replace('Channel Partners','3').str.replace('Enterprise','4').str.replace('Small Business','5')
df['Country'] = df['Country'].str.replace('Canada','1').str.replace('Germany','2').str.replace('France','3').str.replace('Mexico','4').str.replace('United States of America','5')
df['Product'] = df['Product'].str.replace(' Carretera ','1').str.replace(' Montana ','2').str.replace(' Paseo ','3').str.replace(' Velo ','4').str.replace(' VTT ','5').str.replace(' Amarilla ','6')
df['Discount Band'] = df['Discount Band'].str.replace(' None ','1').str.replace(' Low ','2').str.replace(' Medium ','3').str.replace(' High ','4')
````

````python
# merubah data type dari object menjadi integer
df['Segment']  = df['Segment'].astype(int)
df['Country']  = df['Country'].astype(int)
df['Product'] = df['Product'].astype(int)
df['Discount Band'] = df['Discount Band'].astype(int)
````
